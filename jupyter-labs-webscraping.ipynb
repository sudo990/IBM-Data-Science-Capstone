# Web Scraping Falcon 9 and Falcon Heavy Launches Records from Wikipedia
In this lab, we will be performing web scraping to collect Falcon 9 historical launch records from a Wikipedia page.

```
import requests
from bs4 import BeautifulSoup
import pandas as pd

static_url = "https://en.wikipedia.org/w/index.php?title=List_of_Falcon_9_and_Falcon_Heavy_launches&oldid=1027686922"
response = requests.get(static_url)

if response.status_code == 200:
    print("Success! Status Code:", response.status_code)
    
soup = BeautifulSoup(response.text, 'html.parser')
html_tables = soup.find_all('table')
print("Found", len(html_tables), "tables")

# Display a sample of the extracted data
first_launch_table = html_tables[2]
print(first_launch_table.prettify()[:500])
```
