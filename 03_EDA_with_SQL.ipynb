# EDA with SQL
In this assignment, we will understand the Spacex DataSet, load the dataset into the corresponding table in a Db2 database, and execute SQL queries.

```
%load_ext sql
import sqlite3
import pandas as pd

con = sqlite3.connect("my_data1.db")
cur = con.connect()
%sql sqlite:///my_data1.db

# Example Queries
# 1. All Launch Sites
%sql SELECT DISTINCT LaunchSite FROM SPACEXTBL;

# 2. Total Payload Mass
%sql SELECT SUM(PayloadMass) FROM SPACEXTBL WHERE Orbit = 'ISS';

# 3. Maximum Payload Boosters
%sql SELECT Serial, PayloadMass FROM SPACEXTBL WHERE PayloadMass = (SELECT MAX(PayloadMass) FROM SPACEXTBL);
```
